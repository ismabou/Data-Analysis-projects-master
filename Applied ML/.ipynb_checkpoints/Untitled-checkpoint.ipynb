## Part 3 : Use logistic regression to estimate propensity scores for all points in the dataset.

The propensity score of a data point represents its probability of receiving the treatment, based on its pre-treatment features. Therefore, we are not going to take in account re78 as it is after the treatment.

The output in that case is treatt since we need to evaluate the probability of receiving the treatment.

As a slight overfitting is even good for that purpose, we will let the regularization coefficient at 1.

Furthermore, the dataset is small so we are going to use liblinear as solver which means that we must use L1 regularization

```python
from sklearn import linear_model

logistic = linear_model.LogisticRegression(C=1,max_iter=1000,tol=0.0000001,solver='liblinear',penalty='l1')
X_data = [data.loc[i][['age','educ','black','hispan','married','nodegree','re74','re75']].values for i in data.index]
y_data = data['treat'].values
logistic.fit(X_data, y_data)
```

```python
logistic.score(X_data,y_data)
```

Now, for each sample, we will get the probability of receiving the treatment based on the pre-treatment features with the method predict_proba.

```python
propensity_scores = logistic.predict_proba(X_data)
```

## Part 4 : Use the propensity scores to match each data point from the treated group with exactly one data point from the control group, while ensuring that each data point from the control group is matched with at most one data point from the treated group. 
## After matching, you have as many treated as you have control subjects. Compare the outcomes (re78) between the two groups (treated and control). 
## Also, compare again the feature-value distributions between the two groups, as you've done in part 2 above, but now only for the matched subjects. What do you observe? Are you closer to being able to draw valid conclusions now than you were before?

We will create an undirected bipartite graph. The left nodes will correspond to the information of the treatment group and the right nodes will correspond to the information of the control group.
The information we will take in accounts are the matching subjects which are explained by the propensity scores as each score is the probability of having a treatment or not knowing the pre-treatment features;

To do that, we will use networkx package in Python.

The first step will be to create a bipartite graph where the left nodes correspond to the treatment group and the right nodes to the controle group. With that graph, the condition where one treatment group node has to be assigned to exactly one controle group node is satisfied.

The second step is to put an edge between each left and right nodes. The weight of the edge will be one minus the difference between the propensity scores of the nodes as we are using an algorithm that is maximising that sum over the matched pairs.

```python
import networkx as nx
from networkx.algorithms import bipartite

B = nx.Graph()

left_nodes = np.where(y_data == 1)[0] # Nodes for the treatment group
right_nodes = np.where(y_data == 0)[0] # Nodes for the control group

# We add those nodes to the bipartite graph
B.add_nodes_from(left_nodes, bipartite=0) 
B.add_nodes_from(right_nodes, bipartite=1)

'''
Edges to add to the graph
There is an edge between each node on the left and each node on the right
There is not any edges between nodes on the same side
The weight will be designed as follow :
 - Get the propensity score of the node. If it is on the left, we take the propensity score over the treatment group
 - Calculate the absolute value of the difference of the appropriate propensity scores of the nodes
'''
edges_list = []
for left_node in left_nodes:
    for right_node in right_nodes:
        propensity_score_treatment = propensity_scores[left_node][1]
        propensity_score_control = propensity_scores[right_node][0]
        weight = 1 - np.abs(propensity_score_treatment - propensity_score_control)
        edges_list.append((left_node,right_node,weight))
        
B.add_weighted_edges_from(edges_list)
matches = nx.max_weight_matching(B)
```

```python
# The first 185 indexes values are the treatment group.
# Then we take the match of those treatment group

indexes = [matches[i] for i in range(185)] + list(range(185))
```

## Part 5 : Improve your matching by explicitly making sure that you match only subjects that have the same value for the problematic feature. Argue with numbers and plots that the two groups (treated and control) are now better balanced than after part 4.

From the last question, the data has been improved. The only remaining feature that can lead to bad conclusions is black. We will balance the data with the same matching but with an additional requirement : we can only match nodes with the same value on the black feature.

```python
B_condition = nx.Graph()

# We add the nodes with treat = 0 on the left and with treat = 1 on the right
B_condition.add_nodes_from(left_nodes, bipartite=0) 
B_condition.add_nodes_from(right_nodes, bipartite=1)

'''
Edges to add to the graph
There is an edge between each node on the left and each node on the right where the black feature is the same
There is not any edges between nodes on the same side
The weight will be designed as follow :
 - Get the propensity score of the node. If it is on the left, we take the propensity score over the treatment group
 - Calculate the absolute value of the difference of the appropriate propensity scores of the nodes
'''
edges_list_condition = []
for left_node in left_nodes:
    for right_node in right_nodes:
        if data.loc[left_node]['black'] == data.loc[right_node]['black']:
            propensity_score_treatment = propensity_scores[left_node][1]
            propensity_score_control = propensity_scores[right_node][0]
            weight = 1 - np.abs(propensity_score_treatment - propensity_score_control)
            edges_list_condition.append((left_node,right_node,weight))
        
B_condition.add_weighted_edges_from(edges_list_condition)
matches_condition = nx.max_weight_matching(B_condition)
```

```python
# We get the indexes of the treatment group.
# Then we take the match of those treatment group

indexes_condition = [matches_condition[i] for i in matches_condition if i <= 184] + \
                    [i for i in matches_condition if i <= 184]
```

 - Select the parameters we want to tune (n_estimators and max_depth)

### Confusion matrix : It is a matrix which shows the false and true positives over each classes we are trying to predict. In that case, we are going to implement it for the model chosen after the grid search. 

### The confusion matrix will be over the testing set.

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import itertools

def plot_confusion_matrix(cm, classes,
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    """

    print(cm)
    
    plt.figure(figsize=(15,10))
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title('Confusion matrix')
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
    plt.show()
    
    
y_pred = best_random_forest.predict(X_test)
confusion_matrix = confusion_matrix(y_test, y_pred)
class_names = newsgroups['target_names']
plot_confusion_matrix(confusion_matrix, classes=class_names)
```

### The confusion matrix is diagonal. Therefore, the model is giving good results as the true positives and true negatives (true predictions) are much higher than the false positive and false negatives (false predictions)
