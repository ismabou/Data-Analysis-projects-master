# Roadmap :

To get the dataset (ID : tsdec450) we need from eurostat, we can send an HTML request with the right parameters.
After checking the dataset present in eurostat, we will take the unemployment rate for 2016 over all the EU countries.

Definition of unemployment rate used in that case :

Unemployment rates represent unemployed persons as a percentage of the labour force. 
The labour force is the total number of people employed and unemployed. We will take in account the active population. 
Unemployed persons comprise persons aged 15 to 74 who were : 
- without work during the reference week 
- currently available for work (i.e. were available for paid employment or self-employment before the end of the two weeks following the reference week) 
- actively seeking work (i.e. had taken specific steps in the four weeks period ending with the reference week to seek paid employment or self-employment or who found a job to start later within a period of, at most, three months)

We will take all the data within a dataframe with the following schema :
- country

 - Some countries are in the TopoJSON file but not in the dataset

## Preparation of the data and the map ( the answer with the map is in the next section)

```python
eurostat = re.get('http://ec.europa.eu/eurostat/wdds/rest/data/v2.1/json/en/tsdec450?time=2016&unit=PC_POP').text
eurostat = json.loads(eurostat)
```

```python
def get_key_from_dict(dic,value):
    for k,v in dic.items():
        if v == value:
            return k
    return None
```

```python
eurostat_df = pd.DataFrame(columns=['country','country_code','unemployment_rate'])

european_values = ['EA18','EU27','EU28','EA19']

for i in eurostat['value'].keys():
    unemployment_rate = eurostat['value'][i]
    country_code = get_key_from_dict(eurostat['dimension']['geo']['category']['index'],int(i))
    country = eurostat['dimension']['geo']['category']['label'][country_code]
    
    if country == 'Germany (until 1990 former territory of the FRG)':
        country = 'Germany'
    
    if not(country_code in european_values):
        eurostat_df = eurostat_df.append({'country':country,'country_code':country_code,'unemployment_rate':unemployment_rate},ignore_index=True)
        
eurostat_df.head()
```

```python
# We get all the thresholds with the quantiles

quantiles_to_take = [0, 0.2, 0.4, 0.6, 0.8, 1]
quantiles_values = [round(value) for value in map(eurostat_df['unemployment_rate'].quantile,quantiles_to_take)]
```

To plot the map, Folium requires the following :
 - All the countries in the TopoJSON file need to be in the dataframe

```python
# We get all the countries that are in Europe on the topojson file in the following format country_code:country_name
with open('topojson/europe.topojson.json') as jsonfile:
    topology_europe = json.load(jsonfile)
    
countries_topo = {topo['properties']['NAME']:topo['id'] for topo in topology_europe['objects']['europe']['geometries']}

# Based on the country name, we get the countries code that are not present in the dataframe 
# The format is the following : country_code:country_name
missing_countries_name = [country_name for country_name in countries_topo.keys() if country_name not in eurostat_df['country'].values]
```

```python
# For those missing, we will not take them in account in our analysis

for country_name in missing_countries_name:
    eurostat_df = eurostat_df.append({'country':country_name,'country_code':countries_topo[country_name],'unemployment_rate':0},ignore_index=True)
    
# Except for Switzerland
eurostat_df.loc[eurostat_df['country_code'] == 'CH','unemployment_rate'] = 3.3
```

Now that we have collected all the necessary data we will visualise the results. We will use `Folium` to see the results on a map.

The data we have at our disposal is the following :

* `europe.topojson.json` : A topoJSON file which shows __borders between the european countries__.
* `eurostat.p` : A pickle containing the __unemployment rate for each european country__.

We will first get the map of Europe then add the borders. 
At the end, we will make the Choropleth map with the unemployment rates.

```python
# We load the map and center it to Europe

map_eu = folium.Map(location=[50,30],zoom_start=4)
```

```python
# We add the borders of the countries from the topojson file

folium.TopoJson(open('topojson/europe.topojson.json'),
                'objects.europe',
                name='topojson'
               ).add_to(map_eu)
```

```python
# Plot a Choropleth map

# The pale yellow countries will be the one missing
# The legend is based on orange/red to make distinction between the countries with data and without (pale yellow)

# The treshold scale has been taken from the quantiles (0, 0.2, 0.4, 0.6, 0.8, 1)

map_eu.choropleth(
    geo_path='topojson/europe.topojson.json', 
    data=eurostat_df,
    columns=['country', 'unemployment_rate'],
    topojson='objects.europe',
    key_on='feature.properties.NAME',
    threshold_scale=quantiles_values,
    fill_color='OrRd',
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='Unemployment rates (%) in Europe',
    reset=True
)
```

```python
# We then add the marker for Switzerland

folium.Marker(
    [46.92287,8.3829913], # Switzerland center
    popup='CH : ' + str(eurostat_df.loc[eurostat_df['country_code'] == 'CH']['unemployment_rate'].values[0]) + ' %'
).add_to(map_eu)
```

```python
# This function allows us to get the map. It takes as parameters the dataframe, the columns to take in account and title of the legend

def get_map(df,cols,title):
    map_ = folium.Map(location=[46.92287,8.3829913], zoom_start=8)

    folium.TopoJson(open('topojson/ch-cantons.topojson.json'),
                'objects.cantons',
                name='topojson'
               ).add_to(map_)

    map_.choropleth(
        geo_path='topojson/ch-cantons.topojson.json', 
        data=df,
        columns=cols,
        topojson='objects.cantons',
        key_on='feature.id',
        fill_color='YlOrRd',
        fill_opacity=0.7, 
        line_opacity=0.5,
        legend_name=title,
        reset=True
        )
    
    return map_
```

## Preparation of the data for both parts (the next section will present the answers)

```python
# Load dataset for part A

raw_data_q3A = pd.read_excel('data/nationality_metrics.xlsx') 
```

```python
# Prepare the dataframe. To do that, we have to make a pivot of the raw data.

#Get the data for the foreigners
foreigners_data = raw_data_q3A.loc[raw_data_q3A['NationalitÃ©'] == 'Etrangers'][['Canton','Septembre 2017']]
foreigners_data = foreigners_data.rename(columns={'Septembre 2017':'Unemployment rate foreigners'})

#Get the data for the swiss
swiss_data = raw_data_q3A.loc[raw_data_q3A['NationalitÃ©'] == 'Suisses'][['Canton','Septembre 2017']]
swiss_data = swiss_data.rename(columns={'Septembre 2017':'Unemployment rate swiss'})

#Merge the data to get one row for each canton with the unemployment rates of foreigners and swiss as columns
q3A_data = swiss_data.merge(foreigners_data,how='inner',left_on='Canton',right_on='Canton')

# We get the canton code instead of the canton name
for i in range(26):
    q3A_data.set_value(i, 'Canton', canton_to_canton_code[q3A_data.loc[i]['Canton']]) 

#Get the difference between the unemployment rates of foreigners and swiss (the foreigners one is always higher)
q3A_data = q3A_data.convert_objects(convert_numeric=True)
q3A_data['difference'] = q3A_data['Unemployment rate foreigners'] - q3A_data['Unemployment rate swiss']
    
q3A_data.head()
```

```python
# Load dataset for part B

raw_data_q3B = pd.read_excel('data/age_categories.xlsx') 
```

```python
# Prepare the dataframe. To do that, we have again to make a pivot of the raw data.

#This is a dictionary with the translation of the categories from french to english
categories_french_to_english = {'15-24 ans':'15-24 years old','25-49 ans':'25-49 years old','50 ans et plus':'>= 50 years old'}

# This function allows us to get the rows related to a specific category
def get_values_category(df,category,mapping):
    return df.loc[df.Category == category][['Canton','Total']].rename(columns={'Total':'Unemployment rate ' + mapping[category]})

#We select the data of the cantons only
q3B_data = raw_data_q3B.loc[1:78][['Canton','Unnamed: 2','Total']]
q3B_data = q3B_data.rename(columns={'Unnamed: 2':'Category'})

#We get the data for each category with the same schema to prepare the merge
category_1 = get_values_category(q3B_data,'15-24 ans',categories_french_to_english)
category_2 = get_values_category(q3B_data,'25-49 ans',categories_french_to_english)
category_3 = get_values_category(q3B_data,'50 ans et plus',categories_french_to_english)

# We then merge all the categories to get a dataframe with one row for each canton and a column for each category
q3B_data = category_1.merge(category_2,how='inner',left_on='Canton',right_on='Canton')
q3B_data = q3B_data.merge(category_3,how='inner',left_on='Canton',right_on='Canton')

# We get the canton code instead of the canton name
for i in range(26):
    q3B_data.set_value(i, 'Canton', canton_to_canton_code[q3B_data.loc[i]['Canton']]) 

q3B_data = q3B_data.convert_objects(convert_numeric=True)
    
q3B_data.head()
```

The reason why there is such a difference is that the procedure of having a work permit for a foreigner is difficult (Switzerland has quota each year so the selection is difficult) and expensive. Therefore, the companies would rather prefer hire a Swiss than a foreigner.

Due to the last point, the more foreigners you have, the more unemployed you get since the companies will reject a bigger part by rather choosing swiss. 
Therefore, the south/center east of Switzerland has a lower number of foreigners which would result in a lower difference between the unemployment rates of foreigners and swiss.
